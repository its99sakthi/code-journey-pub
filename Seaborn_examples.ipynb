**Plotting Univariate Distribution**

* Function distplot() provides the most convenient way to take a quick look at univariate distribution.
* Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

```
# The Iris data set consists of 50 samples from each of three species of 
# Iris (Iris setosa, Iris virginica and Iris versicolor). Four features 
# were measured from each sample: the length and the width of the sepals 
# and petals, in centimeters. 

# styles available - darkgrid, whitegrid, dark, white, ticks 
sb.set_style("darkgrid") # change styles and see

# palettes available - deep, muted, bright, pastel, dark, colorblind, husl
sb.set_palette(palette='colorblind', n_colors=10)

df = sb.load_dataset('iris')
print(df.head())
sb.distplot(df['petal_length'],kde = False, color='#ff33cc') # , bins=10, , color='yellow'
plt.show()
```

```
# to make it 30% opaque
sb.distplot(df['petal_length'],kde = False,hist_kws={'alpha':0.3})
plt.plot()
```

**Seaborn - Kernel Density Estimates**

* Kernel Density Estimation (KDE) depicts the distribution of the population that the sample beongs to

```
df = sb.load_dataset('iris')
sb.distplot(df['petal_length'],hist=False)
plt.show()
```

```
# showing both
df = sb.load_dataset('iris')
sb.distplot(df['petal_length'])
plt.show()
```

```
# rug plot draws a small vertical tick at each observation to visualize distribution of a single variable. 
# It can be used with either hist or kde or both
df = sb.load_dataset('iris')
sb.distplot(df['petal_length'],rug=True)
plt.show()
```

**Bivariate Distribution** is used to determine the relation between two variables. This mainly deals with relationship between two variables and how one variable is behaving with respect to the other.

* The best way to analyze Bivariate Distribution in seaborn is by using the jointplot() function.

* Jointplot creates a multi-panel figure that projects the bivariate relationship between two variables and also the univariate distribution of each variable on separate axes.

**Scatter Plot**
Scatter plot is the most convenient way to visualize the distribution where each observation is represented in two-dimensional plot via x and y axis.

```
df = sb.load_dataset('iris')
sb.jointplot(x = 'petal_length',y = 'petal_width',data = df)
plt.show()
```

```
df = sb.load_dataset('iris')
sb.jointplot(x = 'petal_length',y = 'petal_width',data = df,kind = 'kde', color='brown')
plt.show()
```

**Visualizing Pairwise Relationship**

* To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. This shows the relationship for combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.

```
df = sb.load_dataset('iris')
sb.set_style("ticks")
sb.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

**Emphasizing continuity with line plots**

* With some datasets, you may want to understand changes in one variable as a function of time, or a similarly continuous variable. In this situation, a good choice is to draw a line plot.

* Draw a single line plot with error bands showing a confidence interval

```
# Functional MRI data
fmri = sb.load_dataset("fmri")
print(fmri.info())
print(fmri.head())
sb.lineplot(x="timepoint", y="signal", data=fmri, ci=90) # remove ci parameter and see confidence interval
```

**Plotting Categorical Data**

* Scatter plots and kde plots which are used to analyze the continuous variables are not suitable when the variable under study is categorical.

* When one or both the variables under study are categorical, we use plots like stripplot()

```
# stripplot() is used when one of the variable under study is categorical. 
# It represents the data in sorted order along any one of the axis.

df = sb.load_dataset('iris')
sb.stripplot(x = "species", y = "petal_length", data = df)
plt.show()
```

**Distribution of Observations**

* Boxplot is a convenient way to visualize the distribution of data through their quartiles.

* Box plots usually have vertical lines extending from the boxes which are termed as whiskers. These whiskers indicate variability outside the upper and lower quartiles i.e., IQR 
* Any Outliers in the data are plotted as individual points.
* The actual box reperesents Q1 to Q3 data

```
sb.boxplot(y = "species", x = "petal_length", data = df) # check with violinplot also
plt.show()
```

**Violin Plots** are a combination of the box plot with the kernel density estimates. So, these plots are easier to analyze and understand the distribution of the data

* The quartile and whisker values from the boxplot are shown inside the violin. As the violin plot uses KDE, the wider portion of violin indicates the higher density and narrow region represents relatively lower density.
* this is a great plot to show data with multiple modes

```
df = sb.load_dataset('tips')
sb.violinplot(x = "day", y = "total_bill", data=df,palette = "husl")
plt.show()
```

* The above plot shows the distribution of total_bill on four days of the week. In addition to that, we can see how the distribution behaves with respect to gender

```
df = sb.load_dataset('tips')
sb.violinplot(x = "day", y = "total_bill",hue = 'sex', data = df)
plt.show()
```

**Bar Plot**
The barplot() shows the relation between a categorical variable and a continuous variable. The data is represented in rectangular bars where the length the bar represents the proportion of the data in that category.

```
# representing 2 fields
df = sb.load_dataset('titanic')
print(df.head())
sb.barplot(x = "sex", y = "survived", data = df)
plt.show()
```

```
# representing 3 fields
df = sb.load_dataset('titanic')
sb.barplot(x = "sex", y = "survived", hue = "class", data = df)
plt.show()
```

**Linear Relationships**
* Most of the times, we use datasets that contain multiple quantitative variables, and the goal of an analysis is often to relate those variables to each other. This can be done through the regression lines.

```
df = sb.load_dataset('tips')
sb.regplot(x = "total_bill", y = "tip", data = df)
plt.show()
```

```
# We can also fit a linear regression when one of the variables takes discrete values
# lmplot can fit linear and also non-linear regression lines
sb.lmplot(x = "size", y = "tip", data = df)
plt.show()
```
